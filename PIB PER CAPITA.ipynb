```python
def regiao (x):
    if x == 'Parana' or x =='Santa Catarina' or x =='Rio Grande do Sul':
        return 'SUL'
    if x== 'Sao Paulo' or x =='Rio de Janeiro' or x =='Minas Gerais' or x =='Espirito Santo':
        return 'SUDESTE'
    if x== 'Rondonia' or x =='Acre' or x =='Para' or x =='Amapa'or x =='Amazonas'or x =='Roraima'or x =='Tocantins':
        return 'NORTE'
    if x== 'Mato Grosso do Sul' or x =='Mato Grosso' or x =='' or x =='Distrito Federal'or x =='Goias':
        return 'CENTRO OESTE'
    return 'NORDESTE'
   
```

```python
df_final['Regiao'] = df_final['Granularidade'].map(regiao)
```

```python
df_final
```

```python
df_final_regiao=df_final.groupby("Regiao").sum()
```
